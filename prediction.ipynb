```python
from tensorflow.keras.models import load_model
import pickle
import pandas as pd 
import numpy as np
```

```python
model = load_model('mymodel.h5')
```

```python
with open('label_encoder_gender.pkl', 'rb') as file:
    label_encoder_gender = pickle.load(file)

with open('onehot_encoder_geo.pkl', 'rb') as file:
    onehot_encoder_geo = pickle.load(file)

with open('scaler.pkl', 'rb') as file:
    scaler = pickle.load(file)
```

```python
input_data = {
    'CreditScore' : 600,
    'Geography' : 'France',
    'Gender' : 'Male',
    'Age' : 40,
    'Tenure' : 3,
    'Balance' : 60000,
    'NumOfProduct':2,
    'HasCrCard' : 1,
    'IsActiveMember' : 1,
    'EstimatedSalary' : 50000
}
```

```python
geo = onehot_encoder_geo.transform([[input_data['Geography']]])
geo_df = pd.DataFrame(geo, columns=onehot_encoder_geo.get_feature_names_out(['Geography']))

input_df = pd.DataFrame([input_data])

input_df['Gender'] = label_encoder_gender.transform([input_data['Gender']])[0]


input_df.rename(columns={'NumOfProduct':'NumOfProducts'}, inplace=True)


input_df = input_df.drop('Geography', axis=1)


input_df = pd.concat([input_df, geo_df], axis=1)


input_df = input_df[scaler.feature_names_in_]


input_scaled = scaler.transform(input_df)


prediction = model.predict(input_scaled)
```

```python
prob = prediction[0][0]

if prob > 0.5:
    print('The customer is likely to churn')
else:
    print('The customer is likely to churn')
```
